```python
import pandas as pd
#Memuat dataset Titanic menggunakan pd.read_csv()
URL = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(URL)

df.head()
```

Q1 Berapa jumlah baris dan kolom pada dataset Titanic? Kolom apa saja yang tersedia?

```python
# Menampilkan jumlah baris dan kolom
shape = df.shape
print(f"Jumlah baris dan kolom: {shape}\n")

# Menampilkan daftar kolom yang tersedia
columns = df.columns
print(f"Kolom yang tersedia: \n{columns.tolist()}")
```

Pada dataset Titanic, terdapat 891 baris dan 12 kolom. Kolom-kolom yang tersedia antara lain:
- 'PassengerId', 'Pclass', 'Name', 'Sex', 'Age', 'SibSp', 'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked', dan 'Survived'.
Kolom-kolom ini memberikan informasi tentang penumpang Titanic, termasuk status keselamatan ('Survived'), kelas tiket ('Pclass'), informasi pribadi seperti usia dan jenis kelamin ('Age' dan 'Sex'), serta data lain yang relevan.

Q2 Kolom mana saja yang memiliki missing values? Berapa jumlah dan proporsinya (%)
per kolom?

```python
# Menampilkan kolom yang memiliki missing values
missing_values = df.isnull().sum()
missing_percent = (df.isnull().sum() / df.shape[0]) * 100

# Menampilkan kolom dengan missing values dan proporsinya
print("Jumlah missing values per kolom:\n", missing_values)
print("\nProporsi missing values per kolom (%):\n", missing_percent)

pd.DataFrame({
    'Jumlah Missing': missing_values,
    'Persentase (%)': missing_percent
})
```

Beberapa kolom dalam dataset Titanic memiliki missing values. Kolom yang paling banyak missing values-nya adalah 'Cabin', yang hampir 77% datanya hilang. Kolom 'Age' juga memiliki sejumlah missing values, yaitu sekitar 20%. Kolom lain seperti 'Embarked' memiliki persentase missing yang lebih kecil.
Persentase missing values ini menunjukkan seberapa besar bagian data yang hilang di setiap kolom, dan ini penting untuk menentukan bagaimana kita akan menangani missing values dalam analisis selanjutnya.

Q3 Berapa persen penumpang yang selamat (Survived=1)? Gunakan operasi NumPy
atau Pandas.

```python
# Menghitung persentase penumpang yang selamat
survived_percentage = (df['Survived'].mean()) * 100
print(f"Persentase penumpang yang selamat: {survived_percentage:.2f}%")
```

Persentase penumpang yang selamat pada Titanic adalah sekitar 38%. Ini dihitung dengan mengambil rata-rata nilai dari kolom 'Survived', di mana 1 menunjukkan bahwa penumpang selamat dan 0 berarti penumpang tidak selamat. Hasil ini memberikan gambaran tentang tingkat kelangsungan hidup penumpang selama kejadian tersebut.

Q4 Filter penumpang wanita dari kelas 1 (Pclass=1). Berapa jumlahnya dan berapa rata- rata usianya?

```python
#Filter penumpang wanita dari kelas 1
female_class_1 = df[(df['Pclass'] == 1) & (df['Sex'] == 'female')]

# Menghitung jumlah dan rata-rata usia
num_female_class_1 = female_class_1.shape[0]
average_age_female_class_1 = female_class_1['Age'].mean()

print(f"Jumlah penumpang wanita di kelas 1: {num_female_class_1}")
print(f"Rata-rata usia penumpang wanita di kelas 1: {average_age_female_class_1:.2f}")
```


Terdapat 94 penumpang wanita dari kelas 1 pada dataset Titanic. Rata-rata usia mereka adalah sekitar 34,6 tahun. Data ini memberikan gambaran tentang demografi penumpang wanita di kelas 1, yang cenderung lebih dewasa dibandingkan dengan kelompok penumpang wanita di kelas lain.

Q5 Gunakan groupby untuk menghitung tingkat keselamatan (mean Survived) per kelas
penumpang. Kelas mana yang paling tinggi?

```python
# Menghitung tingkat keselamatan per kelas penumpang menggunakan groupby
safety_by_class = df.groupby('Pclass').agg(
    Total = ('PassengerId', 'count'),
    Selamat = ('Survived', 'mean')
)
print("Tingkat keselamatan per kelas:\n", safety_by_class)

# Menentukan kelas dengan tingkat keselamatan tertinggi
highest_safety_class = safety_by_class.idxmax()
print(f"Kelas dengan tingkat keselamatan tertinggi: {highest_safety_class}")
```

Tingkat keselamatan per kelas penumpang dihitung dengan mengelompokkan data berdasarkan 'Pclass' dan menghitung rata-rata nilai 'Survived' untuk setiap kelompok. Dari hasil ini, kita menemukan bahwa kelas 1 memiliki tingkat keselamatan tertinggi, yaitu sekitar 62%. Hal ini menunjukkan bahwa penumpang kelas 1 memiliki peluang lebih besar untuk selamat dibandingkan dengan penumpang kelas 2 dan 3.
